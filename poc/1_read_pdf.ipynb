```python
import pymupdf
import os
```

```python
def read_pymupdf(in_path,out_path):
    doc = pymupdf.open(in_path)
    out = open(out_path, "wb") # create a text output
    for page in doc: # iterate the document pages
        text = page.get_text().encode("utf8") # get plain text (is in UTF-8)
        out.write(text) # write text of page
        out.write(bytes((12,))) # write page delimiter (form feed 0x0C)
    out.close()


in_pdfs = ["../data/Ohad Ben-shahar/2501.00836v2.pdf","../data/Ohad Ben-shahar/2021.08.30.458062v1.pdf",
            "../data/Ohad Ben-shahar/1-s2.0-S0042698925000070-main.pdf","../data/Omri Azencot/imagentime.pdf","../data/Omri Azencot/gatlm.pdf"]

in_ix = "pymupdf"

for pd in in_pdfs:
    name = os.path.basename(pd)
    read_pymupdf( pd, f"data/{name}_{in_ix}.txt" )
# 
# "data/read_pdf_output.txt"
```

```python
from pdfminer.high_level import extract_text

def extract_text_pdfminer(in_path,out_path):
    parsed_text =  extract_text(in_path)

    with open(out_path,"w") as f:
        f.write(parsed_text)

in_ix = "pdfminer"

for pd in in_pdfs:
    name = os.path.basename(pd)
    extract_text_pdfminer( pd, f"data/{name}_{in_ix}.txt" )
```
